**Polymorphism**: refers to the ability of an entity ( can be a function/ object) to perform different actions based on
the context
* In Python, polymorphism allows same method, function or operator to behave differently depending on the object it is working with (makes code more flexible and re-usebale )


```python
class Pen: 
    def use(self):
        return "Write"

class Eraser: 
    def use(self):
        return "Erase"


def perform_task(tool):
    return tool.use()


perform_task(Pen())  
```

```python
# ============================================================
# STEP 1: Start with a simple class (NO polymorphism yet)
# ============================================================
# Imagine you only have ONE payment type — Credit Card.
# You'd write something like this:

class CreditCardPayment:
    def process(self, amount):
        print(f"[Credit Card] Validating card... talking to bank API... paid ₹{amount}")

# Use it
cc = CreditCardPayment()
cc.process(500)
```

### Step 3: Build the actual payment types
Each class **inherits** from `PaymentMethod` and **overrides** `process_payment()` with its own logic. Same method name, different behavior — that's polymorphism!

```python
# ============================================================
# STEP 3a: Credit Card — inherits from PaymentMethod
# ============================================================

class CreditCard(PaymentMethod):
    def process_payment(self, amount):
        print(f"[Credit Card] Validating card number...")
        print(f"[Credit Card] Contacting bank API...")
        print(f"[Credit Card] ✅ ₹{amount} paid via Credit Card")

# Test it alone
cc = CreditCard()
cc.process_payment(1500)
```

```python
# ============================================================
# STEP 3b: UPI Payment
# ============================================================

class UPI(PaymentMethod):
    def process_payment(self, amount):
        print(f"[UPI] Redirecting to UPI gateway...")
        print(f"[UPI] ✅ ₹{amount} paid via UPI")

upi = UPI()
upi.process_payment(200)
```

```python
# ============================================================
# STEP 3c: Wallet Payment
# ============================================================

class Wallet(PaymentMethod):
    def __init__(self, balance):
        self.balance = balance

    def process_payment(self, amount):
        if amount > self.balance:
            print(f"[Wallet] ❌ Insufficient balance! Have ₹{self.balance}, need ₹{amount}")
        else:
            self.balance -= amount
            print(f"[Wallet] ✅ ₹{amount} paid. Remaining balance: ₹{self.balance}")

wallet = Wallet(balance=1000)
wallet.process_payment(300)
```

```python
# ============================================================
# STEP 3d: NetBanking Payment
# ============================================================

class NetBanking(PaymentMethod):
    def process_payment(self, amount):
        print(f"[NetBanking] Redirecting to bank login page...")
        print(f"[NetBanking] Authenticating with OTP...")
        print(f"[NetBanking] ✅ ₹{amount} paid via NetBanking")

nb = NetBanking()
nb.process_payment(5000)Q
```

### Step 4: The magic — ONE function handles ALL payment types 🎯
This is where polymorphism pays off. You write a **single function** that accepts ANY payment type. It doesn't care *which* one — it just calls `process_payment()`. 

The right version runs automatically based on the object. **You never need to write `if credit_card... elif upi... elif wallet...`**

```python
# ============================================================
# STEP 4: One function to rule them all
# ============================================================
# This function doesn't know (or care) which payment type it gets.
# It just calls process_payment() — Python figures out which version to run.

def checkout(payment_method, amount):
    print(f"\n--- Processing ₹{amount} ---")
    payment_method.process_payment(amount)
    print("--- Done ---\n")

# Now pass DIFFERENT objects to the SAME function
checkout(CreditCard(), 1500)
checkout(UPI(), 200)
checkout(Wallet(balance=1000), 300)
checkout(NetBanking(), 5000)
```

### Step 5: Why this matters — try adding a new payment type
Suppose tomorrow the business says "Add Crypto payments." Watch how EASY it is — you just add ONE new class. **Nothing else changes.** The `checkout()` function works instantly with it.

```python
# ============================================================
# STEP 5: Adding a new payment type is trivial
# ============================================================

class Crypto(PaymentMethod):
    def process_payment(self, amount):
        print(f"[Crypto] Connecting to blockchain...")
        print(f"[Crypto] ✅ ₹{amount} paid via Bitcoin")

# checkout() works with Crypto WITHOUT any changes!
checkout(Crypto(), 10000)
```

### Summary: What just happened?

| Concept | What it means here |
|---|---|
| **Base class** (`PaymentMethod`) | The contract — "you MUST have `process_payment()`" |
| **Child classes** (`CreditCard`, `UPI`, etc.) | Each implements `process_payment()` differently |
| **Polymorphism** | `checkout()` calls the same method name, but Python runs the correct version based on which object you pass |
| **Why it's powerful** | Adding new payment types = add ONE class. Zero changes to existing code. |

**Without polymorphism:** a giant `if/elif/elif` chain that grows forever.  
**With polymorphism:** one clean function call. Done.

## Abstraction 

```python
import time 
def timer(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        print(f"[TIMER]: {func.__name__} started at {start}")
        elapsed = time.time() - start
        print(f"[TIMER]: {func.__name__} finished in {elapsed:.4f} seconds")
        return func(*args, **kwargs)
    return wrapper


@timer 
def fetch_user_from_db(user_id):
    print("Before sleep")
    time.sleep(1)  # Simulate a slow database query
    print(f"Fetching is done")
    return {"user_id": user_id, "name": "Alice"}


user = fetch_user_from_db(123)
```
